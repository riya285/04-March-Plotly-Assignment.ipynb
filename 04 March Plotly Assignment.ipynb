```python
Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.


import seaborn as sns
import plotly.express as px

# Load the Titanic dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express
scatter_plot = px.scatter(
    titanic_data,
    x='age',
    y='fare',
    title='Scatter Plot of Age vs Fare'
)
scatter_plot.show()








Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.


import plotly.express as px

# Load the tips dataset from Plotly
tips_data = px.data.tips

# Create a box plot using Plotly Express
box_plot = px.box(tips_data, x='day', y='total_bill', title='Box Plot of Total Bill by Day')
box_plot.show()


This code uses the built-in "tips" dataset provided by Plotly Express. It then creates a box plot with the "day" column on the x-axis and the "total_bill" column on the y-axis, showing the distribution of total bills for different days. The show() function is used to display the plot.







Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.


import plotly.express as px

# Load the tips dataset from Plotly
tips_data = px.data.tips

# Create a histogram with pattern shapes and colored by day using Plotly Express
histogram = px.histogram(
    tips_data,
    x='sex',
    y='total_bill',
    color='day',
    pattern_shape='smoker',
    title='Histogram of Total Bill by Sex with Smoker and Day',
    labels={'sex': 'Sex', 'total_bill': 'Total Bill'}
)
histogram.show()



In this code, we're using the px.histogram() function to create a histogram. We're mapping the "sex" column to the x-axis and the "total_bill" column to the y-axis. The "day" column is used to color the bars, and the "smoker" column is used to apply different pattern shapes. The labels parameter is used to customize the axis labels.



Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.



import plotly.express as px

# Load the iris dataset from Plotly
iris_data = px.data.iris

# Create a scatter matrix plot colored by species using Plotly Express
scatter_matrix = px.scatter_matrix(
    iris_data,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    color='species',
    title='Scatter Matrix Plot of Iris Dataset',
    labels={'species': 'Species'}
)
scatter_matrix.update_traces(diagonal_visible=False)  # Hide the histograms on the diagonal
scatter_matrix.show()



Q5. What is Distplot? Using Plotly express, plot a distplot.



distplot was a function provided by the seaborn library that allowed you to plot univariate distributions of observations. It combined a histogram with a kernel density estimate to provide a smooth representation of the distribution. However, starting from Seaborn version 11.0, displot has been introduced to replace distplot and provides a more flexible and consistent way to create distribution plots.

If you're using Plotly Express, you can create a similar distribution plot using the px.histogram() function or the px.histogram_2d() function for 2D histograms. Let's create a basic example using the px.histogram() function:
    
    import plotly.express as px

# Create a sample dataset
data = [1.5, 2, 2, 2.5, 3, 3, 3, 3.5, 3.5, 4, 4, 4, 4.5, 5]

# Create a distribution plot using Plotly Express
dist_plot = px.histogram(data, nbins=10, title='Distribution Plot')
dist_plot.show()




```
